```python
def List_Comprehension():
    squares = [x**2 for x in range(10)]
    print(squares)
    squares = list(map(lambda x: x**2, range(10)))
    print("Lambda:", squares)

    print([(x,y) for x in [1,2,3] for y in [3,1,5] if x != y])
    print([(x,x**2) for x in range(1,6)])

def moreFruits():
    fruits =['orange','apple','pear','banana','kiwi','apple','banana']
    print(fruits.count('apple'))
    print(fruits.count('tangerine'))

    print('Banana -', fruits.index('banana'))

    print ('Banana after 4th item -',fruits.index('banana',4))

    fruits.reverse()
    print('Reversed:',fruits)

    fruits.append('grape')
    print('Appended:',fruits)

    print("Sorted -",end="")
    fruits.sort()
    print(fruits)

    strstrings =sorted("This is a test string from Andrew".split(),key=str.casefold)
    print("Sorted with casefold -",strstrings)

    student_tuples = [
        ('john', 'A', 15), 
        ('jane', 'B', 12), 
        ('dave', 'B', 10)
        ]
    StrStds=sorted(student_tuples, key=lambda student: student[2])
    print("Sorted by age:",StrStds)

def ListAsStack():
    stack =[3,4,5]
    stack.append(6)
    print(stack)
    stack.pop()
    print(stack)
    stack.pop()
    stack.pop()
    print(stack," ")

def ListAsQueue():
    from collections import deque
    queue = deque(["Eric","John","Michael"])
    queue.append("Terry")
    queue.append("Graham")
    print(queue)
    queue.popleft()
    queue.popleft()
    print(queue)

#List_Comprehension()
#moreFruits()
#ListAsStack()
ListAsQueue()
```
